```python
from mendeleev import get_table
import pandas as pd
```

```python
data = get_table('elements')
series = get_table('series')
groups = get_table('groups')

data = pd.merge(data, series, left_on='series_id', right_on='id',
                    how='left', suffixes=('', '_series'))
data = pd.merge(data, groups, left_on='group_id', right_on='group_id',
                    how='left', suffixes=('', '_group'))
```

```python
from bokeh.models.mappers import LinearColorMapper, CategoricalColorMapper, LogColorMapper
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool, ColumnDataSource, FixedTicker
from bokeh.models import LogColorMapper, BasicTicker, ColorBar
```

```python
output_notebook()
```

```python
fig = figure(plot_width=800,
             plot_height=600,
             tools='pan,box_zoom,resize,save',
             toolbar_location='above',
             toolbar_sticky=False)

xname = 'atomic_number'
yname = 'covalent_radius_pyykko'

lcm = LinearColorMapper(palette='Viridis256', low=data[yname].min(), high=data[yname].max())

fig.circle(x=xname, y=yname, fill_alpha=1.0, size=10,
           source=ColumnDataSource(data=data))#, fill_color={'field': yname, 'transform': lcm},
          #line_color={'field': yname, 'transform': lcm})

color_bar = ColorBar(color_mapper=lcm, ticker=BasicTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))

fig.add_layout(color_bar, 'right')

show(fig)
```
